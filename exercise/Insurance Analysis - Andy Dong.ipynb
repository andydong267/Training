```python
print('Female: Percent of Sample + Average Charge')
print( len(data[ data['sex'] == 'female']) / len(data['sex']) )
print( data[ data['sex'] == 'female']['charges'].mean())

print()

print('Male: Percent of Sample + Average Charge')
print( len(data[ data['sex'] == 'male']) / len(data['sex']) )
print( data[ data['sex'] == 'male']['charges'].mean())

print()

print('Ratio for Female:Male')
print((len(data[ data['sex'] == 'female']) / len(data['sex'])) / (len(data[ data['sex'] == 'male']) / len(data['sex'])))
print( data[ data['sex'] == 'female']['charges'].mean() / data[ data['sex'] == 'male']['charges'].mean())
```

```python
print('Smoker: Percentage, Mean, Median')
print( (data['smoker'] == 'yes').sum() / len(data['smoker']) )
print( data[(data['smoker'] == 'yes')]['charges'].mean() )
print( data[(data['smoker'] == 'yes')]['charges'].median() )

print()

print('Non-Smoker: Percentage, Mean, Median')
print( (data['smoker'] == 'no').sum() / len(data['smoker']) )
print( data[(data['smoker'] == 'no')]['charges'].mean() )
print( data[(data['smoker'] == 'no')]['charges'].median() )
```

```python
# Grouping by age, charges is moderately correlated to bmi as age goes up.
# 

# Grouped by smoker status: being a non-smoker, charges and age are moderately correlated (0.627947)
#                           being a smoker, charges and bmi are strongly correlated (0.806481)
#                                           charges and age are weakly correlated (0.368224)

# Grouped by gender: only one percent difference between male and female population
#                    but female charges are 10% less than males

# Hypothesis: You get charged more if you are male vs female. Smoker vs non-smoker. 
# As a non-smoker, age is more strongly correlated with charges.
# As a smoker, bmi is strongly correlated with charges with age having a weak correlation
```
